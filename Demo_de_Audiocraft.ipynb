```python
from audiocraft.models import musicgen
from audiocraft.utils.notebook import display_audio
import torch
```

Asegurate de activar en el Runtime T4 GPU
Entra al menu Runtime > Change runtime type y selecciona "T4 GPU"

```python
model = musicgen.MusicGen.get_pretrained('medium', device='cuda')
model.set_generation_params(duration=8)
```

# Describe tu musica aquí
Cada elemento se transformará en un clip de audio

```python
res = model.generate([
    'synthwave relax with a smooth beat'
],
progress=True)
display_audio(res, 32000)
```
